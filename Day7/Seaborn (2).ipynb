# '''The Python visualization library Seaborn is based on matplotlib and provides a high-level interface for drawing attractive statistical graphics.'''

```python
#The basic steps to creating plots with Seaborn are:
# 1. Prepare some data
# 2. Control figure aesthetics
# 3. Plot with Seaborn
# 4. Further customize your plot

import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()

```

```python
sns.set_style("whitegrid")
g = sns.lmplot(x="tip",y="total_bill",data=tips,aspect=2)
g = (g.set_axis_labels("Tip","Total bill (USD)").
set(xlim=(0,10),ylim=(0,100)))
plt.title("title")
plt.show(g)
```

```python
#Seaborn also offers built-in data sets:
import pandas as pd
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
titanic.head()
iris.head(100)
#iris['species'].value_counts()
```

```python
#Scatterplot,Scatterplot with one,categorical variable
sns.stripplot(x="species",y="petal_length",data=iris)
```

```python
#Categorical scatterplot with  non-overlapping points
sns.swarmplot(x="species", y="petal_length", data=iris)
```

```python
sns.pairplot(data=iris)
```

```python
#Bar Chart Show point estimates and confidence intervals with scatterplot glyphs
print(titanic.head())
sns.barplot(x="sex",y="survived", hue="class", data=titanic)
```

```python
#Count Plot  Show count of observations
print(titanic['deck'].value_counts())
sns.countplot(x="deck",data=titanic,palette="Greens_d")
```

```python
#Point Plot Show point estimates and  confidence intervals as rectangular bars
sns.pointplot(x="class", y="survived",hue="sex", data=titanic,palette={"male":"g","female":"m"},markers=["^","o"],linestyles=["-","--"])
```

```python
#Boxplot 
sns.boxplot(x="alive",y="age",hue="adult_male",data=titanic)
```

```python
sns.boxplot(data=iris,orient="h")  #Boxplot with wide-form data
```
