```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('diabetes.csv')
data.info()
data.head()
```

```python
data.describe()
```

```python
# Box plots for each feature
plt.figure(figsize=(12, 10))
for i, column in enumerate(data.columns, 1):
    if column == 'Outcome':
        continue
    plt.subplot(3, 3, i)
    sns.boxplot(y=data[column])
    plt.ylabel(column)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, column in enumerate(data.columns, 1):
    plt.subplot(3, 3, i)
    data[column].hist(bins=20, edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
data.corr()
```

```python
heatmap = sns.heatmap(data.corr(), annot=True)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x=data['Outcome'], y=data['Pregnancies'], ci='sd', hue= 'Outcome')
plt.title('Average Number of Pregnancies by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Average Number of Pregnancies')
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, column in enumerate(data.columns, 1):
    if column == 'BMI' or column == 'Outcome':
        continue
    plt.subplot(3, 3, i)
    plt.bar(data['BMI'], data[column])
    plt.xlabel('BMI')
    plt.ylabel(column)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 10))
for i, column in enumerate(data.columns, 1):
    if column == 'BMI':
        continue
    plt.subplot(3, 3, i)
    sns.stripplot(x=data['BMI'], y=data[column], data=data)
    plt.xlabel('BMI')
    plt.ylabel(column)
plt.tight_layout()
plt.show()
```

```python
# Box plot of Age by Outcome
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Outcome', y='Age')
plt.title('Age Distribution by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Age')
plt.show()
```

```python
# Scatter plot of Glucose vs Age colored by Outcome

sns.scatterplot(data=data, x='Age', y='Glucose', hue='Outcome', alpha=0.7)
plt.title('Scatter Plot: Glucose vs Age colored by Diabetes Outcome')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.legend(['Non-Diabetic', 'Diabetic'], loc='upper right')
plt.show()
```

```python
# Pairplot of BMI, SkinThickness, and Outcome
sns.pairplot(data, vars=['BMI', 'SkinThickness','Age'], hue='Outcome')
plt.suptitle('Pairplot of BMI, SkinThickness by Diabetes Outcome', y=1.02)
plt.show()
```

```python
# Pairplot of all numerical features by Outcome
sns.pairplot(data, hue='Outcome')
plt.suptitle('Pairplot of All Features by Diabetes Outcome', y=1.02)
plt.show()
```

```python
# Violin plot of Glucose distribution by Outcome
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='Outcome', y='Glucose')
plt.title('Distribution of Glucose Levels by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Glucose')
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
# Box plot of DiabetesPedigreeFunction by Outcome
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Outcome', y='DiabetesPedigreeFunction')
plt.title('Diabetes Pedigree Function by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Diabetes Pedigree Function')
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```
